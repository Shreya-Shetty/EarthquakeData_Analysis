                                            UNDERSTANDING THE DATASET
time-Time when the event occurred.                                                                                             
latitude-Decimal degrees latitude. Negative values for southern latitudes.                                                     
longitude-Decimal degrees longitude. Negative values for western longitudes.                                                   
depth-The depth where the earthquake begins to rupture(Depth of the event in kilometers)                                       
mag-The magnitude for the event.                                                                                               
magType-The method or algorithm used to calculate the preferred magnitude for the event.                                       
nst-The total number of seismic stations used to determine earthquake location.                                                 
gap-The largest azimuthal gap between azimuthally adjacent stations (in degrees)                                               
dmin-Horizontal distance from the epicenter to the nearest station                                                             
rms- The parameter provides a measure of the fit of the observed arrival times to the predicted arrival times for this location net-The ID of a data contributor.                                                                                              
id-eight-character network-assigned code.	                                                                                   
updated-Time when the event was most recently updated.                                                                         
place-Textual description of named geographic region near to the event.                                                        
type-Type of seismic event                                                                                                     
horizontalError-Uncertainty of reported location of the event in kilometers.                                                   
depthError-Uncertainty of reported depth of the event in kilometers.                                                           
magError-The estimated standard error of the magnitude.                                                                         
magNst-The total number of seismic stations used to calculate the magnitude for this earthquake.                               
status-Indicates whether the event has been reviewed by a human                                                                 
locationSource-The network that originally authored the reported location of this event.                                       
magSource-Network that originally authored the reported magnitude for this event.                                               

<font size="5">Checking the need for data cleaning.<font>

```python
#checking the presence of null values in each column
eq_data.isnull().sum()
```

```python
tot_missingdata=eq_data.isnull().sum().sum()
tot_data=eq_data.shape[0]*eq_data.shape[1]
print("Number of missing data: ",tot_missingdata)
percentage_of_missing_data=tot_missingdata/tot_data*100
print("Percentage of missing data: ",percentage_of_missing_data)
```

Since the data missing is only 5% and the data in each row is independent, we cannot fill the missing data. 

```python
eq_data.describe()
```

The data seems to be consistent as the standard deviation is less concerning the title of the column. Lesser standard deviation indicates that the values are closer to the mean value, which shows consistency in the data.

```python
#eq_data.isnull().sum() gives the total number of rows having NaN
#eq_data.shape[0] gives the total number of rows present
#if the above mentioned vales are equal, we get output as 'True'.This means the column is full of NaN vales, which is not useful for our analysis
eq_data.isnull().sum()==eq_data.shape[0]
```

Since there are no columns with all the values as NaN, we don't have to delete any columns. <font size=3>And the data seems to be good for analysis now.

<font size=5>Analyzing the missing data to find if there are any patterns in it.

```python
eq_data.groupby('locationSource')['magError'].sum()
```

Since the locationSource 'ak' has 0 magnitude error, let us print the data for further analysis.

```python
eq_data[eq_data['locationSource']=='ak']['magError']
```

This data seems to be ambiguous, let's see if there are any other columns with all null values.

```python
#Counting number of null values in the selected columns
eq_data[eq_data['locationSource']=='ak'][['nst','gap','dmin','horizontalError']].isnull().count()
```

```python
sum=eq_data[eq_data['locationSource']=='ak'][['magError','nst','gap','dmin','horizontalError']].isnull().count().sum()
print(sum/tot_missingdata*100,"% of missing data is from locationSource 'ak'")
```

<font size='4'>Observation 1: 67% of the missing data is from locationSource 'ak'. From this data, we can conclude that there must either be a problem with the network or the organization recording these values does not need the errors.

<font size='5'> Analyzing the data

```python
#Finding the number of networs listed
unique_networks=eq_data['magSource'].unique()
print("Number of Networks:",len(unique_networks),"\nList of Networks:",unique_networks)
```

<font size='4'> Observation 2: There are 13 networks in total.

```python
#Calculating the mean depth of the earthquakes occured, grouped by their reported location
eq_data.groupby('locationSource')['depth'].mean().sort_values(ascending=False)
```

<font size=4>Observation 3: It can be inferred that the location network 'us' has detected heavy earthquakes as its mean value is the highest. It can be said that the network 'us' is the strongest and fastest in the list. Also, when the dataset is closely observed we can say that 'us' network is the largest as it covers places from all over the world like New Zeland, Chile, China, and Alaska. 

```python
#Counting the number of earthquakes detected by every network
eq_data['magSource'].value_counts()
```

<font size='4'>Observation 4: We can see that network 'ci' reports the highest number of earthquakes from the dataset.

<font size='5'>Plotting graphs to have better visualizations

```python
#Finding the deepest earthquake and intensity of eathquakes in general
eq_data['depth'].plot()
plt.ylabel('Depth(in Kms)',fontsize=12)
plt.xlabel('Serial number(in the dataframe)',fontsize=12)
plt.title('Depth of earthquakes')
```

<font size='4'> Observation 5:
From the above visualization, we can see:                                                                                       
~The deepest earthquake was 600+ Kms below sea level.                                                                           
~Most of the earthquakes had a crack of less than 100 Kms.                                                                     

```python
#Finding the severity of the earthquakes listed
eq_data['mag'].hist(bins=1000)
plt.ylabel('Count',fontsize=12)
plt.xlabel('Magnitude',fontsize=12)
```

<font size='4'> Observation 6:
It can be inferred that most of the earthquake's magnitude ranged between 0.4 and 1.5, which are minor earthquakes that are generally not felt and neglected by a human.

```python
#Plotting the magnitude error to see if the networks are in a good condition.
ser=eq_data['magError']
plt.scatter(ser.index, ser)
plt.ylabel('Magnitude(of the eorrors)',fontsize=12)
plt.xlabel('Serial number(in the data frame)',fontsize=12)
plt.title('Scattered plot to analyze the magnitudes recorded')
```

<font size='4'> Observation 7:
From the above plot, we can see that there is a maximum error of 0.57 and most of the errors lay between 0.0 and 0.3 magnitude, which is an optimum level of error. We can conclude that the networks are working properly.

```python
#Using the bar graph to see the proportion of human reviewed earthquakes
_=eq_data['status'].value_counts().plot(kind="bar")
_=plt.ylabel('Count',fontsize=12)
```

<font size='4'> Observation 8: Earthquakes are mostly detected automatically. We can see that only around 33% of the earthquakes are reviewed by humans.
